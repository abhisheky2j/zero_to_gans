## Function 1 - torch.Tensor.view(*shape)

Returns a new tensor called view tensor, view tensor shares the same storage with the original tensor, view tensor's shape can be different from original tensor.

Example 1: Create a new torch.Tensor object with different size

```python
t2 = t.view(4,4)
t2
```

Example 2: Verify storage sharing feature

```python
t2[0,0] = 1
```
